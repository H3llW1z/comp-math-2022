## Основные теоретические положения

ддитивная цепочка для $ n \in \mathbb{N} $  - последовательность натуральных чисел, от единицы, где каждый последующий элемент равен сумме каких-то двух предыдущих. 

$ a_i = a_j + a_k, \quad k \le j \le i, \quad i = 1, 2, ..., r $

$ l(n) = r $ - наименьшая длина $r$ для которой сущ. аддитивная цепочка для $ n $ (ф-я от $n$)

Неравенство для метода множ.:

$ l(mn) \leq l(m) + l(n) $

m-арный метод: 

$ \quad l(n) \le m - 2 + (k + 1)t, \quad$ 

Бин. метод:

$l(n) \leq \lambda(n) + \nu(n) - 1, \quad$

Пара $ (j, k), 0 \le k \le j < i $ называется шагом $i$

Виды и свойства шагов:

Виды:
* удвоение: $ j = k = i - 1 $
* звездный: $ j = i - 1 $
* малый шаг: $ \lambda(a_i) = \lambda(a_{i-1}) $

Свойства:
* Шаг 1 - всегда удвоение
* Удвоение - звездный шаг
* За удвоением всегда следует звездный шаг
* Если $ i $-й шаг не малый, то ($i+1$)-й шаг либо малый, либо зведный, либо и тот, и другой
* Если ($i+1$)-й шаг ни звездный, ни малый, то $i-й$ шаг должен быть малым.




Если аддитивная цепочка включ. $d$ и $f = r - d$ неудвоений, то 

$n \le 2^{d-1} F_{f+3}$, где $F_j$ - число Фиббоначи

Если аддитивная цепочка включает $f$ неудвоений и $S$ малых шагов, то

${S \le f \le} {S \over {1 - lb(\varphi)}}$,где $\varphi = {{\sqrt{5} + 1} \over 2} $ - золотое сечение

### Метод брауэра

```python
def find_brauer_chain(n, k, chain):
    if(n < 2**k):
        for i in range(1, 2**k):
            chain.append(i)
        return
    
    elif(n >= 2**k):
        q = n // (2**k) 
        find_brauer_chain(q, k, chain)
        for i in range(1, k+1):
            q *= 2
            chain.append(q)
        chain.append(n)
```

### Алгоритм дробления вектора индексов для нахождения минимальной звёздной цепочки

```python
from math import log, ceil, floor


def min_m(n):
    return ceil(log(n, 2))

def max_m(n):
    return floor(log(n, 2)) + bin(n).count('1')


def reduce_vector(vec, l_index, r_index):
    i = r_index
    while vec[i] == 1 and i >= l_index:
        i -= 1
    if i < l_index:
        return False
    vec[i] = vec[i] - 1
    for j in range(i + 1, r_index + 1):
        vec[j] = j + 1

    return True


def generate_chain_by_index_vector(index_vector):
    res = [1]
    for i in index_vector:
        res.append(res[-1] + res[i - 1])
    return res


def create_chain(n):
    if n < 3:
        return list(range(1, n + 1))

    for m in range(min_m(n), max_m(n)):
        full_vector = list(range(1, m + 1))
        q = m // 2
        full_vector[q - 1] += 1
        while reduce_vector(full_vector, 0, q - 1):
            chain = generate_chain_by_index_vector(full_vector)
            if chain[m] == n:
                return chain
            elif chain[q] + m - q <= n <= chain[q] * (2 ** (m - q)):
                while reduce_vector(full_vector, q, m - 1):
                    chain = generate_chain_by_index_vector(full_vector)
                    if chain[m] == n:
                        return chain
                for i in range(q, m):
                    full_vector[i] = i + 1
```

```python
import time


def test(num):
    start = time.time()

    res_chain = create_chain(num)

    stop = time.time()

    print(f'Время работы алгоритма для числа {num} = {round(stop - start, 3)} сек')
    print(f'Итоговая цепочка: {" ".join(map(str, res_chain))}')
    print(f'Длина цепочки = {len(res_chain)}')
    
```
